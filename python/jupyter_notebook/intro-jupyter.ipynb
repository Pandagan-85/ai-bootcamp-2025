## Comandi Base
Possiamo usare dei comandi speciali come:

* `!ls` per vedere il contenuto della directory (come da terminale)
* I comandi magici **vanno messi come prima cosa nella cella**:
    * che iniziano con singolo `%` per `%timeit`per misurare la velocita del parametro che passiamo
    * che iniziano con doppio `%%` per `%%timeit` misura tutto il codice contenuto in tutta la cella non solo la stringa
    * `%run main.py` serve per far girare un file chiamato `main.py`
* 

```python
!ls
```

```python
obj = range(10)
```

```python
%timeit [num for num in obj]
```

```python
%%timeit
for num in obj:
    num
```

## Esercizio di gruppo

fare il plot dal csv dato

```python
import pandas as pd
import matplotlib.pyplot as plt

df2 = pd.read_csv("https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/refs/heads/master/dati/soggetti-guariti.csv")

df2
```

```python
df2.head()
```

```python
print(df2.columns)
```

```python
# salvo il dataframe in un csv
df2.to_csv('data/soggett-guariti-covid.csv', index=False)
```

```python
# Plot per vedere il numero di guariti senza somm. per regione

fig,ax = plt.subplots(figsize=(12, 8))
ax.bar(df2["reg"], df2["guariti_senza_somm"])

# Etichette e titolo
ax.set_xlabel('Regione')
ax.set_ylabel('Guariti senza somministrazione')
ax.set_title('Guariti senza somministrazione per Regione')

# Ruota le etichette sull'asse X per migliorare la visibilità
plt.xticks(rotation=35)

plt.show();
```

```python
# Crea la figura con due sottotrame affiancate (1 riga, 2 colonne)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 8))

ax1.bar(df2["reg"], df2["guariti_senza_somm"], color='blue')

# Etichette e titolo
ax1.set_xlabel('Regione')
ax1.set_ylabel('Guariti senza somministrazione')
ax1.set_title('Guariti senza somministrazione per Regione')
ax1.tick_params(axis='x', rotation=90)  # Ruota etichette di 90 gradi
ax1.set_ylim([0, 700])  # Imposta la scala dell'asse y


ax2.bar(df2["reg"], df2["guariti_post_somm"], color='orange')
# Etichette e titolo
ax2.set_xlabel('Regione')
ax2.set_ylabel('Guariti con somministrazione')
ax2.set_title('Guariti con somministrazione per Regione')
ax2.tick_params(axis='x', rotation=90)   # Ruota etichette di 90 gradi
ax2.set_ylim([0, 700])  # Imposta la scala dell'asse y


plt.tight_layout() 
```
